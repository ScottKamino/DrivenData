```python
%matplotlib inline

import matplotlib.pyplot as py 
import pandas as pd
import numpy as np
import sklearn 
```

```python
# Load mapping from restauarant ids to yelp ids
id_map = pd.read_csv("Data/restaurant_ids_to_yelp_ids.csv")
id_dict = {}
```

```python
id_map.head()
```

```python
# Each Yelp ID may correspond to up to 4 Boston IDs
# So create a dictionary with Yelp IDs being keys and Boston IDs being values.
# Much of this code block borrowed from driven data blog.
for i, row in id_map.iterrows():
    # get the Boston ID
    boston_id = row["restaurant_id"]
    
    # get the non-null Yelp IDs
    non_null_mask = ~pd.isnull(row.ix[1:])
    yelp_ids = row[1:][non_null_mask].values
    
    for yelp_id in yelp_ids:
        id_dict[yelp_id] = boston_id
```

```python
with open("Data/yelp_boston_academic_dataset_updated/yelp_academic_dataset_review.json", 'r') as review_file:
    # the file is not actually valid json since each line is an individual
    # dict -- we will add brackets on the very beginning and ending in order
    # to make this an array of dicts and join the array entries with commas
    review_json = '[' + ','.join(review_file.readlines()) + ']'

# read in the json as a DataFrame
reviews = pd.read_json(review_json)

# drop columns that we won't use
reviews = reviews.drop(['review_id', 'type'], axis=1)

# replace yelp business_id with boston restaurant_id
map_to_boston_ids = lambda yelp_id: id_dict[yelp_id] if yelp_id in id_dict else np.nan
reviews.business_id = reviews.business_id.map(map_to_boston_ids)
```

```python
reviews.head()
```

```python
# Let's unpack the votes column and create columns for funny_votes, useful_votes, and cool_votes.
reviews['funny_votes'] = np.nan
reviews['useful_votes'] = np.nan
reviews['cool_votes'] = np.nan

funny = []
useful = []
cool = []

for x in reviews.votes:
    funny.append(x['funny'])
    useful.append(x['useful'])
    cool.append(x['useful'])

reviews['funny_votes'] = pd.Series(funny)
reviews['useful_votes'] = pd.Series(useful)
reviews['cool_votes'] = pd.Series(cool)
                  
reviews = reviews.drop(['votes'], axis=1)
```

```python
# rename first column to restaurant_id so we can join with boston data
reviews.columns = ["restaurant_id", "date", "stars", "text", "user_id", "funny_votes", "useful_votes", "cool_votes"]    

# drop restaurants not found in boston data
reviews = reviews[pd.notnull(reviews.restaurant_id)]

print reviews.shape
reviews.head()
```

```python
# Get number of one star, two star, three star, four star, and five star reviews.
one_star = reviews[reviews['stars']==1]
two_star = reviews[reviews['stars']==2]
three_star = reviews[reviews['stars']==3]
four_star = reviews[reviews['stars']==4]
five_star = reviews[reviews['stars']==5]

one_star = one_star.groupby('restaurant_id')
two_star = two_star.groupby('restaurant_id')
three_star = three_star.groupby('restaurant_id')
four_star = four_star.groupby('restaurant_id')
five_star = five_star.groupby('restaurant_id')

one_star = one_star['stars'].count()
two_star = two_star['stars'].count()
three_star = three_star['stars'].count()
four_star = four_star['stars'].count()
five_star = five_star['stars'].count()

one_star = pd.DataFrame({'restaurant_id': pd.Series(one_star.index),
                        'one_star': pd.Series(one_star.values)}, columns = ['restaurant_id', 'one_star'])

two_star = pd.DataFrame({'restaurant_id': pd.Series(two_star.index),
                        'two_star': pd.Series(two_star.values)}, columns = ['restaurant_id', 'two_star'])

three_star = pd.DataFrame({'restaurant_id': pd.Series(three_star.index),
                        'three_star': pd.Series(three_star.values)}, columns = ['restaurant_id', 'three_star'])

four_star = pd.DataFrame({'restaurant_id': pd.Series(four_star.index),
                        'four_star': pd.Series(four_star.values)}, columns = ['restaurant_id', 'four_star'])

five_star = pd.DataFrame({'restaurant_id': pd.Series(five_star.index),
                        'five_star': pd.Series(five_star.values)}, columns = ['restaurant_id', 'five_star'])

star_df = pd.merge(pd.merge(pd.merge(pd.merge(one_star, two_star, how='outer'), three_star, how='outer'), four_star, how='outer'), \
            five_star, how='outer')

star_df.head()
```

```python
# Fill NaNs with 0s
star_df['one_star'] = star_df['one_star'].fillna(0)
star_df['two_star'] = star_df['two_star'].fillna(0)
star_df['three_star'] = star_df['three_star'].fillna(0)
star_df['four_star'] = star_df['four_star'].fillna(0)
star_df['five_star'] = star_df['five_star'].fillna(0)
```

```python
# Group by restaurant to get average rating.
restaurant = pd.groupby(reviews, by='restaurant_id')

avg_stars = pd.Series(restaurant['stars'].mean())

avg_stars = pd.DataFrame({'restaurant_id': pd.Series(avg_stars.index),
                          'avg_stars': pd.Series(avg_stars.values)}, columns = ['restaurant_id' ,'avg_stars'])
```

```python
# Merge restauarant data with data_df
star_df = pd.merge(star_df, avg_stars, how='left')
```

```python
# Read in Boston health inspection data
train_labels = pd.read_csv("Data/trainingLabels.csv", index_col=0)
submission = pd.read_csv("Data/SubmissionFormat.csv", index_col=0)
train_labels = train_labels.reset_index()

# Drop date and id columns
train_labels = train_labels.drop(['date', 'id'], axis=1)

# Reset index
train_labels.head()
```

```python
# For each restauarant, get average *, **, and *** violations
violations = train_labels.groupby('restaurant_id')
minor = violations['*'].mean()
medium = violations['**'].mean()
major = violations['**'].mean()

minor = pd.DataFrame({'restaurant_id' : pd.Series(minor.index),
                      '*': pd.Series(minor.values)}, columns = ['restaurant_id', '*'])

medium = pd.DataFrame({'restaurant_id' : pd.Series(minor.index),
                      '**': pd.Series(medium.values)}, columns = ['restaurant_id', '**'])

major = pd.DataFrame({'restaurant_id' : pd.Series(minor.index),
                      '***': pd.Series(major.values)}, columns = ['restaurant_id', '***'])

violations = pd.merge(pd.merge(minor, medium, how='outer'), major, how='outer')

violations['*'] = violations['*'].fillna(0)
violations['**'] = violations['**'].fillna(0)
violations['***'] = violations['***'].fillna(0)
```

```python
restaurant_df = pd.merge(star_df, violations, on=['restaurant_id'])
output_df = restaurant_df[['restaurant_id', '*', '**', '***']]
restaurant_df = restaurant_df.drop(['*', '**', '***'], axis=1)
restaurant_df.tail(n=10)
```

```python
# Take text of all reviews.
# Merge all reviews for a single restaurant
# Create corpus for all reviews

def restaurant_reviews(restaurant_df, reviews_df):
        reviews_dictionary = {}
        N = len(reviews)
        
        for i, (pid, row) in enumerate(restaurant_df.iterrows()):
            restaurant_reviews = reviews_df[reviews_df['restaurant_id'] == row['restaurant_id']]
            all_text = ' '.join(restaurant_reviews.text)
            reviews_dictionary[row['restaurant_id']] = all_text
        
        return pd.Series(reviews_dictionary)
```

```python
review_text = restaurant_reviews(restaurant_df, reviews)
```

```python
# Now let's create a model!
X = data_df.drop('restaurant_id', axis=1)
X = X.as_matrix().astype(np.float)
```

```python
# Set up outcome arrays
output_df.head()
y_small = output_df['*']
y_medium = output_df['**']
y_large = output_df['***']
```

```python
# Set up K-fold cross-validation
from sklearn.cross_validation import KFold

def run_cv(X, y, clf_class, **kwargs):
    kf = KFold(len(y), n_folds=5, shuffle=True)
    y_pred = y.copy()
    
    for train_index, test_index in kf:
        X_train, X_test = X[train_index], X[test_index]
        y_train = y[train_index]
        clf = clf_class(**kwargs)
        clf.fit(X_train, y_train)
        y_pred[test_index] = clf.predict(X_test)
    return y_pred
```

```python
from sklearn.ensemble import RandomForestRegressor as RF

def accuracy(y_true, y_pred):
    return np.mean((y_true-y_pred)**2)

print "Random Forest:"
print "One Star violations:"
print "%.3f" %accuracy(y_small, run_cv(X, y_small, RF))

print "Random Forest:"
print "Two Star violations:"
print "%.3f" %accuracy(y_medium, run_cv(X, y_medium, RF))

print "Random Forest:"
print "Three Star violations:"
print "%.3f" %accuracy(y_large, run_cv(X, y_large, RF))
```
